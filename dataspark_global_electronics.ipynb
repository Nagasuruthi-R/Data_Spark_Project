```python
#Import necessary librarires

import pandas as pd
from datetime import timedelta
```

```python

def process_customers():
    customers = pd.read_csv(r"C:\Users\Green Gen Tech\Downloads\Customers_data.csv", encoding='ISO-8859-1')
    print(f"raw_data:\n{customers.head()}\n")
    customers['Birthday'] = pd.to_datetime(customers['Birthday']).dt.date
    customers.drop('State Code', axis=1, inplace=True)
    customer_cleaned = customers.copy()
    customer_cleaned_path = r"C:\\Users\\Green Gen Tech\\Downloads\\Customer_data_cleaned.xlsx"
    customers.to_excel(customer_cleaned_path, index=False)
    print(f"Processed Customers:\n{customers.head()}\n")

def process_stores():
    stores = pd.read_csv(r"C:\Users\Green Gen Tech\Downloads\Stores_data.csv", encoding='ISO-8859-1')
    print(f"raw_data:\n{stores.head()}\n")
    stores['Open Date'] = pd.to_datetime(stores['Open Date'], errors='coerce')
    stores['Square Meters'] = stores['Square Meters'].fillna(0)
    stores_cleaned_path = r"C:\\Users\\Green Gen Tech\\Downloads\\stores_data_cleaned.xlsx"
    stores.to_excel(stores_cleaned_path, index=False)
    print(f"Processed Stores:\n{stores.head()}\n")

def process_exchange_rates():
    exchange_rate = pd.read_csv(r"C:\Users\Green Gen Tech\Downloads\Exchange_Rates_data.csv", encoding='ISO-8859-1')
    print("Initial DataFrame:")
    print(exchange_rate.head())
    print("Initial Data Types:")
    print(exchange_rate.dtypes)

    exchange_rate['Date'] = pd.to_datetime(exchange_rate['Date'], errors='coerce')
    print("DataFrame After Conversion:")
    print(exchange_rate.head())
    print("Data Types After Conversion:")
    print(exchange_rate.dtypes)

    excel_file_path = r"C:\\Users\\Green Gen Tech\\Downloads\\exchange_rates_data_cleaned.xlsx"
    exchange_rate.to_excel(excel_file_path, index=False)

    exchange_rate_loaded = pd.read_excel(excel_file_path)
    print("DataFrame After Loading from Excel:")
    print(exchange_rate_loaded.head())
    print("Data Types After Loading from Excel:")
    print(exchange_rate_loaded.dtypes)

def process_products():
    product = pd.read_csv(r"C:\Users\Green Gen Tech\Downloads\Products_data.csv")

    unit_cost_usd = []
    for value in product['Unit Cost USD']:
        if isinstance(value, str):
            # Remove '$', commas, and strip any spaces before converting to float
            unit_cost_usd.append(float(value.replace('$', '').replace(',', '').strip()))
        else:
            unit_cost_usd.append(value)
    product['Unit Cost USD'] = unit_cost_usd

    unit_price_usd = []
    for value in product['Unit Price USD']:
        if isinstance(value, str):
            # Remove '$', commas, and strip any spaces before converting to float
            unit_price_usd.append(float(value.replace('$', '').replace(',', '').strip()))
        else:
            unit_price_usd.append(value)
    product['Unit Price USD'] = unit_price_usd

    product['Unit Cost USD'].fillna(0, inplace=True)
    product['Unit Price USD'].fillna(0, inplace=True)

    product_cleaned_path = r"C:\\Users\\Green Gen Tech\\Downloads\\products_data_cleaned.xlsx"
    product.to_excel(product_cleaned_path, index=False)
    print(f"Processed Products:\n{product.head()}\n")


def process_sales():
    sales = pd.read_csv(r"C:\Users\\Green Gen Tech\Downloads\Sales_data.csv", encoding='ISO-8859-1')
    print(f"raw_data:\n{sales.head()}\n")
    sales['Order Date'] = pd.to_datetime(sales['Order Date'], errors='coerce')
    sales['Delivery Date'] = pd.to_datetime(sales['Delivery Date'], errors='coerce')
    
    sales['Difference'] = (sales['Delivery Date'] - sales['Order Date']).dt.days
    average_difference = sales['Difference'].mean()

    sales['Delivery Date'].fillna(sales['Order Date'] + pd.to_timedelta(average_difference, unit='days'), inplace=True)
    sales.drop(columns=['Difference'], inplace=True)

    sales_cleaned_path = r"C:\\Users\\Green Gen Tech\\Downloads\\Sales_data_cleaned.xlsx"
    sales.to_excel(sales_cleaned_path, index=False)
    print(f"Processed Sales:\n{sales.head()}\n")

def process_all_data():
    process_customers()
    process_stores()
    process_exchange_rates()
    process_products()
    process_sales()

# Run the function to process all data
process_all_data()
```
