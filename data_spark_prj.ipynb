```python
import pandas as pd

# Read the CSV file
product = pd.read_csv("C:\\Users\\Green Gen Tech\\Downloads\\Products.csv")

# Function to remove the '$' and convert to float, with error handling
def convert_to_float(value):
    try:
        return float(value.replace('$', ''))
    except ValueError:
        return None  # or you can return np.nan if you want to handle missing values later

# Apply the conversion function to the relevant columns
product['Unit Cost USD'] = product['Unit Cost USD'].apply(convert_to_float)
product['Unit Price USD'] = product['Unit Price USD'].apply(convert_to_float)
product['Unit Cost USD'].fillna(0, inplace=True)
product['Unit Price USD'].fillna(0, inplace=True)
# Copy the cleaned DataFrame
product_cleaned = product.copy()

# Save the cleaned DataFrame to a new Excel file
output_path = r"C:\\Users\\Green Gen Tech\\Downloads\\products_cleaned.xlsx"
product_cleaned.to_excel(output_path, index=False)

# Display the updated DataFrame
print(product.head())
```
